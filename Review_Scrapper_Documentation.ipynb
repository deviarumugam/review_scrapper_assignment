For Web review scrapping assignment, I select the "WALMART" retail showroom website. There are thousands of products available for review scrapping. For assignment purpose, i selected few products. I scrapped the reviews of this products. The product ID of the products are given to search the products review. The product ID's are extracted from the Walmart site itself. All the reviews are scrapped for the searchable product. Positive reviews and negative reviews are scrapped from the overall reviews. Star ratings are scrapped from the reviews.

                                    *WALMART REVIEW SCRAPPING*
                                     -SCRAPPED OVERALL REVIEWS OF THE PRODUCT
                                    -POSITIVE REVIEWS
                                    -NEGATIVE REVIEWS
                                    -STAR RATINGS 
                                    
In Jupyter notebook, using Beautiful Soap I scrapped the reviews from Walmart. The screenshots are given below. I must deploy the code in cloud, so I want to change my jyputer notebook code into PyCharm. In PyCharm, I developed the code which I was used in the jyputer notebook. For deployment, I used Heroku.

PyCharm Steps:

1.	Create app.py. Main code is available here.
2.	Create requirements.txt. 
3.  Create static and template folder for UI purpose. All html and css files are available here.
4.  Create proc and gunicorn file.
5.  Run pip freeze requirements.txt.
6.  Pip install gunicorn.
7.  Change the host name.

Heroku Deployment steps:

1.  Login to Heroku-create new app-named as Walmartreviewpage.
2.  Download and install the Heroku CLI.
3.  heroku login
4.  heroku git:clone -a walmartreviewpage
5.  cd walmartreviewpage
6.  git add .
7.  git commit -am "make it better"
8.  git push heroku master

After Deployment, we got
https://walmartreviewpage.herokuapp.com/
